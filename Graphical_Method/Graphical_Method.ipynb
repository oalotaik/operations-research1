## Graphical Method

In this notebook, the graphical method for solving linear programs is presented (using 2 variables). 

**Example 1**

Consider the following two-variable linear program:

Objective function: $$ Max \ z = 3x + 5y $$
Constraints: $$  x + 2y \leq 6 \\ 2x + 3y \leq 10 \\ x + y \leq 5 \\ x \leq 4 \\ y \leq 3 \\ x, y \geq 0 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from intvalpy import lineqs
%matplotlib inline
```

```python
x = np.linspace(0, 20, 2000)
```

**Each inequality determines a unique half-plane**

We graph the first constraint: $ x + 2y \leq 6 $
> **Tip**: draw equations for the inequalities, i.e. set $\leq, or \geq$ as $ = $. For each equation find two points by **1-** setting x = 0 and then finding y (0, y), and **2-** setting y = 0 and then finding x (x, 0). Now you have two points (0, y) and (x, 0). Connect them with a line.

```python
# x + 2y ≤ 6
y1 = 3 - (x/2)
plt.plot(x, y1, label=r'$x + 2y\leq6$');
plt.xlim((0, 7));
plt.ylim((0, 6));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now we graph the second constraint: $ 2x + 3y \leq 10 $

```python
# 2x+ 3y ≤ 10
y2 = (10/3) - 2*(x/3)
plt.plot(x, y1, label=r'$x + 2y\leq6$');
plt.plot(x, y2, label=r'$2x + 3y\leq10$');
plt.xlim((0, 7));
plt.ylim((0, 6));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now we graph the third constraint: $ x + y \leq 5 $
> **Note:** this constraint is a redundant constraint because it does not make the feasible region smaller.

```python
y3 = 5 - x

plt.plot(x, y1, label=r'$x + 2y\leq6$');
plt.plot(x, y2, label=r'$2x + 3y\leq10$');
plt.plot(x, y3, label=r'$x + y\leq5$');
plt.xlim((0, 7));
plt.ylim((0, 6));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now we graph: $ x \leq 4 $ and $ y \leq 3 $

```python
plt.plot(x, y1, label=r'$x + 2y\leq6$');
plt.plot(x, y2, label=r'$2x + 3y\leq10$');
plt.plot(x, y3, label=r'$x + y\leq5$');
plt.axvline(x=4, color='r', label=r'$x\leq4$');
plt.axhline(y=3, color='k', label=r'$y\leq3$');
plt.xlim((0, 7));
plt.ylim((0, 6));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

**Feasible Region:** the region that satisfies all the inequalities

Now we graph the feasible region and find the extreme points

```python
A = -np.array([[1, 2],
               [2, 3],
               [1, 1],
               [1, 0],
               [0, 1],
               [-1, 0],
               [0, -1]])
b = -np.array([6, 10, 5, 4, 3, 0, 0])

print('Extreme points are:')
lineqs(A, b, title='Feasible Region', color='gray',
       alpha=0.2, s=5, size=(7,6), save=False, show=True)
```

**Example 2**

Consider this LP: 

Objective: $$ Max \ Z = 5x + 3y $$

Constraints: $$ 3x + 7y \leq 21 \\ 5x + 3y \leq 15 \\ -2x + 3y \leq 6 \\ x, y \geq 0 $$

We draw the first constraint: $3x + 7y \leq 21$

```python
# 3𝑥+7𝑦≤21
y1 = 3 - 3*(x/7)
plt.plot(x, y1, label=r'$3x + 7y \leq 21$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

We draw the second constraint: $5x + 3y \leq 15$

```python
# 5𝑥+3𝑦≤15
y2 = 5 - 5*(x/3)
plt.plot(x, y1, label=r'$3x + 7y \leq 21$');
plt.plot(x, y2, label=r'$5x + 3y \leq 15$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Next, draw the constraint: $-2x + 3y \leq 6$

```python
# −2𝑥+3𝑦≤6
y3 = 2 + 2*(x/3)
plt.plot(x, y1, label=r'$3x + 7y \leq 21$');
plt.plot(x, y2, label=r'$5x + 3y \leq 15$');
plt.plot(x, y3, label=r'$-2x + 3y \leq 6$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now we draw the **feasible region** and find the **extreme points**

```python
A = -np.array([[3, 7],
               [5, 3],
               [-2, 3],
               [-1, 0],
               [0, -1]])
b = -np.array([21, 15, 6, 0, 0])

print('Extreme points are:')
lineqs(A, b, title='Feasible Region', color='gray',
       alpha=0.2, s=5, size=(6,3), save=False, show=True)
```

**Example 3**

Consider this LP: 

Objective: $$ Max \ z= 3x + 4y $$

Constraints: $$ 6x + 7y \leq 42 \\ 2x \geq 4 \\ -x + y \geq 5 \\ x, y \geq 0 $$

We draw the first constraint: $6x + 7y \leq 42$

```python
# 6𝑥+7𝑦≤42
y1 = 6 - 6*(x/7)
plt.plot(x, y1, label=r'$6x + 7y \leq 42$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now draw: $2x \geq 4$

```python
# 2𝑥≥4
plt.plot(x, y1, label=r'$6x + 7y \leq 42$');
plt.axvline(x=2, color='r', label=r'$2x\geq4$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now: $-x + y \geq 5$

```python
# −𝑥+𝑦≥5
y2 = 5 + x
plt.plot(x, y1, label=r'$6x + 7y \leq 42$');
plt.axvline(x=2, color='r', label=r'$2x\geq4$');
plt.plot(x, y2, label=r'$-x + y \geq 5$');
plt.xlim((0, 7.5));
plt.ylim((0, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

```python
A = -np.array([[6, 7],
               [-2, 0],
               [1, -1],
               [-1, 0],
               [0, -1]])
b = -np.array([42, -4, -5, 0, 0])

print('Extreme points are:')
lineqs(A, b, title='Feasible Region', color='gray',
       alpha=0.2, s=5, size=(6,3), save=False, show=True)

print('This LP is infeasible')
```

> **Note:** the above LP is **infeasible**

**Example 4**

Consider this LP:

Objective: $$ Max \ z = 5x + 11y $$

Constraints: $$ 3x - y \leq 6 \\ 8x + 3y \geq 24 \\ y \geq 5 \\ x, y \geq 0 $$

Draw the first constraint: $3x - y \leq 6$

```python
# 3𝑥−𝑦≤6
y1 = 3*x - 6
plt.plot(x, y1, label=r'$3x - y \leq 6$');
plt.xlim((0, 7.5));
plt.ylim((-7, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Now: $8x + 3y \geq 24$

```python
# 8𝑥+3𝑦≥24
y2 = 8 - 8*(x/3)
plt.plot(x, y1, label=r'$3x - y \leq 6$');
plt.plot(x, y2, label=r'$8x + 3y \geq 24$');
plt.xlim((0, 7.5));
plt.ylim((-7, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

Finally: $y \geq 5$

```python
# 𝑦≥5
plt.plot(x, y1, label=r'$3x - y \leq 6$');
plt.plot(x, y2, label=r'$8x + 3y \geq 24$');
plt.axhline(y=5, color='r', label=r'$y \geq 5$');
plt.xlim((0, 7.5));
plt.ylim((-7, 7.5));
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```

```python
A = -np.array([[6, 7],
               [-2, 0],
               [1, -1],
               [-1, 0],
               [0, -1]])
b = -np.array([42, -4, -5, 0, 0])

print('Extreme points are:')
lineqs(A, b, title='Feasible Region', color='gray',
       alpha=0.2, s=5, size=(6,3), save=False, show=True)
```
